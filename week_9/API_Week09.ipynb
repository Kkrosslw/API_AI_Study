## 词法分析

```python
def baidu_lexer(text):
    """百度API词法分析"""
    url = "https://aip.baidubce.com/rpc/2.0/nlp/v1/lexer?&access_token="+token
    headers = {
        "Content-Type":"application/json"
    }
    data = {   
        "text":text
    }
    encode_data = json.dumps(data).encode("UTF-8")
    return requests.post(url,headers=headers,data=encode_data)
```

```python
r = baidu_lexer("中山大学是一所双一流高校")
r.content
r_str = str(r.content,encoding="GBK")
r_dict = json.loads(r_str)
pd.json_normalize(r_dict["items"])
```

## DNN语言模型
中文DNN语言模型接口用于输出切词结果并给出每个词在句子中的概率值,判断一句话是否符合语言表达习惯。

```python
def baidu_DNN(text):
    """百度API词法分析"""
    url = "https://aip.baidubce.com/rpc/2.0/nlp/v2/dnnlm_cn?&access_token="+token
    headers = {
        "Content-Type":"application/json"
    }
    data = {   
        "text":text
    }
    encode_data = json.dumps(data).encode("UTF-8")
    return requests.post(url,headers=headers,data=encode_data)
```

```python
c = baidu_DNN("水中月是天上月，眼前人是心上人")
c_str = str(c.content,encoding="GBK")
c_str
c_dict = json.loads(c_str)
# pd.json_normalize(c_dict["items"])
c_dict

# ppl代表此句话的通畅度,值越低代表句子越通顺
```

## 依存句法分析
依存句法分析接口可自动分析文本中的依存句法结构信息，利用句子中词与词之间的依存关系来表示词语的句法结构信息（如“主谓”、“动宾”、“定中”等结构关系），并用树状结构来表示整句的结构（如“主谓宾”、“定状补”等）。

```python
def baidu_anylize(text,mode=1):
    """百度API词法分析"""
    url = "https://aip.baidubce.com/rpc/2.0/nlp/v1/depparser?&access_token="+token
    headers = {
        "Content-Type":"application/json"
    }
    data = {   
        "text":text
    }
    encode_data = json.dumps(data).encode("UTF-8")
    return requests.post(url,headers=headers,data=encode_data)
```

```python
d = baidu_anylize("你吃饭了吗")
d_str = str(d.content,encoding="GBK")
d_dict = json.loads(d_str)
d_dict
```
